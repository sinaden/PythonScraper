```python




def scrape_it(my_url):
    global row

    #my_url = 'http://webuyit.dk/modules/product/index.php?id=29'


    # opening up connection, grabbing the page
    uClient = uReq(my_url)
    page_html = uClient.read()
    uClient.close()


    page_soup = soup(page_html, "html.parser")

    list_item = page_soup.findAll("a",{"href":"#product_spec"})

    for i1 in range(0, len(list_item)):
        print("*********************************")
        item = list_item[i1]
        img_title = item.findAll("span", {"class":"image-title"})
        brand = img_title[0].text.strip()
        box = list_item[i1]
        str_1 = box["onclick"]
        st_in = str_1.find("product_detail_")
        en_in = str_1.find("').toggle()");
        detail_id = str_1[st_in:en_in]
        hidden_container = page_soup.findAll("div",{"id":detail_id})
        phones = hidden_container[0].findAll("span", {"class":"phone-title"})
        memories = []
        links = []
        for i2 in range(0, len(phones)):
            
            arr[0].append(' ')
            arr[1].append(' ')
            arr[2].append(' ')
            arr[3].append(' ')
            arr[4].append(' ')
            arr[5].append(' ')

            arr[0].append(' ')
            arr[1].append(' ')
            arr[2].append(' ')
            arr[3].append(' ')
            arr[4].append(' ')
            arr[5].append(' ')

            arr[0].append(' ')
            arr[1].append(' ')
            arr[2].append(' ')
            arr[3].append(' ')
            arr[4].append(' ')
            arr[5].append(' ')
            
            arr[0].append(' ')
            arr[1].append(' ')
            arr[2].append(' ')
            arr[3].append(' ')
            arr[4].append(' ')
            arr[5].append(' ')
            
            phone = phones[i2]
            title = phone.text.strip()
            print(title)
            st_in = title.find(brand) + len(brand) + 1
            memory = title[st_in:].strip()

            arr[1][row] = brand
            arr[1][row + 1] = brand
            arr[1][row + 2] = brand
            arr[1][row + 3] = brand

            print('model:', brand)
            
            arr[2][row] = memory
            arr[2][row + 1] = memory
            arr[2][row + 2] = memory
            arr[2][row + 3] = memory

            print('memory: ', memory)
            memories.append(memory)
            link = phone.a['href']
            
            arr[4][row] = link
            arr[4][row + 1] = link
            arr[4][row + 2] = link
            arr[4][row + 3] = link

            links.append(link)
            find_price_and_condition(link)
            
            row += 4
            print('- - - - - - - -  - \n')
```

```python
def find_price_and_condition (url) :

    print("url:" , url)
    uClient = uReq(url)
    page_html = uClient.read()
    uClient.close()
    page_soup = soup(page_html, "html.parser")
    
    the_brand = page_soup.findAll("a",{"href":"javascript:void(0)"})
    print('brand:', the_brand[1].text)
    
    arr[0][row] = the_brand[1].text
    arr[0][row + 1] = the_brand[1].text
    arr[0][row + 2] = the_brand[1].text
    arr[0][row + 3] = the_brand[1].text

    

    list_perfekt = page_soup.findAll("span",{"id":"price_1"})
    
    
    print('perfekt:', list_perfekt[0].text.strip()[4:])
    arr[3][row] = 'perfekt'
    
    arr[5][row] = list_perfekt[0].text.strip()[4:]
    
    list_god = page_soup.findAll("span",{"id":"price_6"})
    print('god:', list_god[0].text.strip()[4:])
    
    arr[3][row + 1] = 'god'
    
    arr[5][row + 1] = list_perfekt[0].text.strip()[4:]
    
    list_skadet = page_soup.findAll("span",{"id":"price_7"})
    print('skadet:', list_skadet[0].text.strip()[4:])
    
    
    arr[3][row + 2] = 'defekt'
    arr[3][row + 3] = 'rimelig'
    
    arr[5][row + 2] = list_perfekt[0].text.strip()[4:]
    arr[5][row + 3] = list_perfekt[0].text.strip()[4:]


```
